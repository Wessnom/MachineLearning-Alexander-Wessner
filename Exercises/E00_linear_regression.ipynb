## E0.0 A) Simulate Phone Dataset

**Dataset variables:** <br>
x: Call minutes <br>
y: SEK/month

simulate 400 x-values from the r.v. $X \sim \mathcal{N}(100, 100)$ <br>
This line specifies the generation of 400 random values from a normal distribution with a mean (μ) of 100 and a standard deviation (σ) of 100.

```python
# random seed for reproducibility
np.random.seed(42)

# 400 random x-values normally distributed with my = 100, sigma = 100
X = np.abs(np.random.normal(100, 100, 400))

# 400 random normally distributed noise points with mu = 0, sigma = 50
epsilon = np.random.normal(0, 50, 400)

beta_0 = 25 #intercept
beta_1 = 2 # slope

# y as a function of X with random noise added
y = beta_1 * X + beta_0 + epsilon
```

```python
sns.scatterplot( x = X, y = y, label = "Datapoints")
plt.axhline(y = 0, color = 'orange', linestyle = '--')
plt.axvline(x = 300, color = "green", linestyle = "--")

plt.legend();
```

## E0.0 B) Remove Outliers

```python
# getting indexes where time is above 300
itemindex = np.where(X > 300)

# times above 300
X[itemindex]

X = np.clip(X, 0, 300) # clip values to be within 0 to 300
y = np.maximum(y, 0) # ensure y is always positive
```

```python
sns.regplot(x = X, y = y, label = "Datapoints", line_kws={"color": "black", "label": "Based on entire Dataset"})
sns.lineplot(x = X, y = 2 * X + 25, color = "red", label = "Ground Truth Line")
plt.legend();
```
